```python
# importing required libraries
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
df = pd.read_excel("MLZilla Data Set.xlsx")
df
```

```python
print(df.shape, df.columns)
```

```python
target = 'Churn'
```

```python
df = df.replace([' ',''], np.nan)  # Convert empty strings to NaN
```

```python
if 'customerID' in df.columns:
  df= df.drop(columns = ['customerID'])
#df.info()
```

```python
missing_values = (df.isna().sum())
missing_values.values.sum() / (len(df)*len(df.columns)) * 100
```

```python
missing_df = (df.isna().sum() / len(df) * 100).to_frame(name='Missing(%)').assign(Type=df.dtypes)
missing_df
```

```python
for col in missing_df.index:
    #Replace Categorical Data wuth less then 5% missing vales with mode
    if df[col].nunique() in range(2, 5) and col != target and 0 < missing_df.loc[col, 'Missing(%)'] < 5:
        if not df[col].mode().empty:  # Check if mode exists
            df[col] = df[col].fillna(df[col].mode()[0])  # Explicit assignment
    #Replace Numerical Data wuth less then 5% missing vales with mean
    elif col != target and 0 < missing_df.loc[col, 'Missing(%)'] < 5 and str(missing_df.loc[col, 'Type']) != 'object':
        df[col] = df[col].fillna(df[col].mean())  # Explicit assignment

# Recalculate missing_df after filling
missing_df = (df.isna().sum() / len(df) * 100).to_frame(name='Missing(%)').assign(Type=df.dtypes)
missing_df = missing_df[missing_df['Missing(%)'] > 0]  # Keep only columns with missing values

missing_df
```

```python
#Replacing Tenure, Monthly Charges, Total Charges with Mean since they are continuous data
df['tenure'] = df['tenure'].fillna(df['tenure'].median())
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(df['MonthlyCharges'].mean())
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
## 4.2 Handling Columns with 5% - 15% Missing Values
# Impute if important, drop if not critical
important_cols = ['InternetService', 'LastInteractionDate']
less_important_cols = ['PromotionalOffersUsed', 'CustomerSatisfactionScore']
```

```python
missing_df = (df.isna().sum() / len(df) * 100).to_frame(name='Missing(%)').assign(Type=df.dtypes)
missing_df
```

```python
for col in missing_df.index:
    #Replace Categorical Data wuth less then 5% missing vales with mode
    if df[col].nunique() in range(2, 5) and col != target and 0 < missing_df.loc[col, 'Missing(%)'] < 5:
        if not df[col].mode().empty:  # Check if mode exists
            df[col] = df[col].fillna(df[col].mode()[0])  # Explicit assignment
    #Replace Numerical Data wuth less then 5% missing vales with mean
    elif col != target and 0 < missing_df.loc[col, 'Missing(%)'] < 5 and str(missing_df.loc[col, 'Type']) != 'object':
        df[col] = df[col].fillna(df[col].mean())  # Explicit assignment

# Recalculate missing_df after filling
missing_df = (df.isna().sum() / len(df) * 100).to_frame(name='Missing(%)').assign(Type=df.dtypes)
missing_df = missing_df[missing_df['Missing(%)'] > 0]  # Keep only columns with missing values

missing_df
```

```python
#Replacing Tenure, Monthly Charges, Total Charges with Mean since they are continuous data
df['tenure'] = df['tenure'].fillna(df['tenure'].median())
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(df['MonthlyCharges'].mean())
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
## 4.2 Handling Columns with 5% - 15% Missing Values
# Impute if important, drop if not critical
important_cols = ['InternetService', 'LastInteractionDate']
less_important_cols = ['PromotionalOffersUsed', 'CustomerSatisfactionScore']
```

```python
#Replacing Important Columns with Mode
for col in important_cols:
    if df[col].isna().sum() > 0:
        mode_value = df[col].mode()
        if not mode_value.empty:
            df[col] = df[col].fillna(mode_value.iloc[0])  # Use `.iloc[0]`
```

```python
#Deleting Less Import Important columns
for col in less_important_cols:
  if col in df.columns:
    df.drop(col, axis=1, inplace=True)
```

```python
#removing the rows where the target is null
df = df[df['Churn'].notna()]
```

```python
print("\nFinal Missing Values Check:")
print((df.isna().sum() / len(df)) * 100)
```

```python
#CHECKING THE DUPLICACY
duplicate_rows = df[df.duplicated()]
print("Duplicate Rows:")
print(duplicate_rows)
print("Duplicates Before Removal:", df.duplicated().sum())
```

```python
df= df.drop_duplicates()
```

```python
print("Duplicates After Removal:", df.duplicated().sum())
```

```python
# Descriptive Statistics
desc_stats = df.describe().round(2)
desc_stats

#Defining Range And Max People Falling in range
desc_stats.loc['Range'] = desc_stats.loc['min'].astype(str) + " - " + desc_stats.loc['max'].astype(str)
desc_stats.loc['MAX'] = desc_stats.loc['25%'].astype(str) + " - " + desc_stats.loc['75%'].astype(str)

desc_stats.columns = ['SeniorCitizen','Tenure(months)', 'MonthlyCharges($)', 'TotalCharges($)',
       'CustomerLifetimeValue($)', 'NetPromoterScore', 'AverageCallDuration(minutes)',
       'NumberOfComplaints', 'BillingCycle(days)', 'NumberOfDependents']
print("Summary Of Numerical Variables")
desc_stats

```
